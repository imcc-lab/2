#### Meta Characters

```python
import re

# []
a="my name is sakshi, my age is 22, class MCA"
b=re.findall("[0-9]",a)
print(b)
```

```python
# ^ caret in bracket is compliment
a="my name is sakshi, my age is 22, class MCA"
#excludes
b=re.findall("[^0-9]",a)
print(b)
```

```python
# ^ caret without []
a="my name is sakshi, my age is 22, class MCA"
#checks for given characters avaiable at first or not
b=re.findall("^my",a)
print(b)
c=re.findall("^sakshi",a)
print(c)
```

```python
# $ - checks whether characters are available at last or not
a="my name is sakshi, my age is 22, class MCA"
b=re.findall("A$",a)
print(b)
c=re.findall("sakshi$",a)
print(c)
```

```python
# . - symbol
a="my name is sakshi, my age is 22, class MCA"
#printing any character between a and e i.e. for each word - one character between a & e
b=re.findall("a.e",a)
print(b)
#printing any character between a and e i.e. for each word - two characters between a & e
c=re.findall("n..e",a)
print(c)
d=re.findall("s....i",a)
print(d)
```

```python
# | or symbol - also known as pipe symbol
a="my name is sakshi, my age is 22, class MCA"
b=re.findall("is | my",a)
print(b)
c=re.findall("vedika | sakshi",a)
print(c)
```

```python
# ? - 0 or 1 occurence
a="my name is sakshi"
a2="my name is sashi"
a3="my name is saaaashi"
# ? mark's left side character is available in my string so i will get the output
# even if the character is not avialable i will get the output
b=re.findall("sak?shi",a)
print(b)
c=re.findall("sak?shi",a2)
print(c)
d=re.findall("sa?shi",a3)
print(d)
```

```python
# + - 1 or more occurence
a="my name is sakshi"
a2="my name is sashi"
a3="my name is saaaashi"
b=re.findall("sak+shi",a)
print(b)
c=re.findall("sak+shi",a2)
print(c)
d=re.findall("sa+shi",a3)
print(d)
```

```python
# * - 0 or more occurence
a="my name is sakshi"
a2="my name is sashi"
a3="my name is saaaashi"
b=re.findall("sak*shi",a)
print(b)
c=re.findall("sak*shi",a2)
print(c)
d=re.findall("sa*shi",a3)
print(d)
```

```python
# {} - braces
a="my name is sakshi, my age is 22, class MCA"
# my is available minimum 1 (once) or maximum 2(twice)
b=re.findall("is{1,2}",a)
print(b)
```

##### Common Patterns and techniques

```python
# email validation
a="my email id is sakshi.kokare@gmail.com"
b=re.findall(r"[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-z]{2,}", a)
print(b)
```

```python
# validate a phone number
a="my phone number is 1234567890 and her is 7410258"
b=re.findall(r"\b\d{10}\b",a)
print(b)
```

```python
text="Python is very powerful programming language"
a=re.findall(r"\bp\w+",text)
print(a)
```

#### Lookahead and Lookbehind techniques

```python
import re
# positive look ahead (?=...)
# Matches the words that are followed by "ing" but doesn't include "ing"
text="I am running, eating and playing"
result=re.findall(r"\w+(?=ing)",text)
print(result)
```

```python
#Negative Lookahead (?!...)
text="cat dog catalog category"
result=re.findall(r"cat(?!alog)",text)
print(result)
```

```python
#Positive lookbehind
text="rate: Rs30, Profit: Rs2000, Loss:500"
result=re.findall(r"(?<=Rs)\d+",text)
print(result)
```

```python
#Negative Lookbehind
text="rate: Rs30, Profit: Rs2000, Loss:500, RI:78"
result=re.findall(r"(?<!Rs)\d+",text)
print(result)
```

```python
#Combining Lookahead and Lookbehind
#extracting username before @
text="Email: info.imcc@mespune.in, test.mastersoft@mesp.com"
a=re.findall(r"(?<=\b)\w+(?=@)",text)
print(a)
```

#### Building Regex Patterns

```python
# Phone  number validation (10 digits)
text="call us on 1234567890 or 9876543210 or 45678913"
a=re.findall(r"(\b\d{10}\b)",text)
print(a)
```

```python
#Extracting Dates from sentences(DD/MM/YYYY)
text="12/12/2025 is celebrated as anti-corruption day and yesterday 11/12/2025 was International Animal Rights Day"
a=re.findall(r"\b\d{2}/\d{2}/\d{4}\b",text)
print(a)
```

```python
# extracting hastags
text="The subjects are #Python, #OB and #Java"
a=re.findall(r"#\w+",text)
print(a)
```

```python
#Strong password Regex
#Regex:Atleast 8 characters, one uppercase, 1 lowercase, 1 digit, 1 special character
pattern=re.compile(r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@!$%&*?])[A-Za-z\d@!$%&*?]{8,}$')
passwords=["Password@123","Pass12","PasswordMax1456","MaxPassword@123"]
for pwd in passwords:
    if pattern.match(pwd):
        print(f"{pwd} is valid")
    else:
        print(f"{pwd} is invalid")
    
```

```python
# URL validation
pattern=re.compile(
    r'^(https?|ftp)://'    #protocol in url 's' is optional
    r'([A-Za-z0-9.-]+)'    #domain name
    r'([A-Za-z]{2,})'      # sub domain or top level domain
    r'(:\d+)?'              # port number if any that is optional
    r'(/\[^\s])*?$'          #optional path or query if any '\s' anything excluding the whitespace
)

urls=[
    "https://www.example.com",
    "http://imcc.edu.in",
    "ftp://files.server.net:12/path/to/file",
    "https://example.com/page?query=abc"
    "invalid_url"
]

for u in urls:
    if pattern.match(u):
        print(f"{u} is valid URL")
    
```
